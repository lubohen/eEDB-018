```python
import numpy as np
import matplotlib.pyplot as plt

# Dados fictcios para a espectrografia de cada instrumento, visando destacar suas caractersticas nicas
frequencia = np.linspace(0, 5000, 500)

# Espectrografia do Trompete
trompete = (np.sin(frequencia * 0.25) * np.exp(-(frequencia - 2000)**2 / 500000)) + \
           (np.sin(frequencia * 0.25) * np.exp(-(frequencia - 2500)**2 / 500000))

# Espectrografia da Bateria
bateria = np.random.normal(0, 0.5, 500) * np.exp(-frequencia / 1000)

# Espectrografia do Contrabaixo
contrabaixo = np.sin(frequencia * 0.02) * np.exp(-frequencia / 1000)

# Plotar os grficos
plt.figure(figsize=(15, 9))

# Trompete
plt.subplot(3, 1, 1)
plt.plot(frequencia, trompete, color='gold')
plt.title('Espectrografia do Trompete')
plt.ylabel('Amplitude')

# Bateria
plt.subplot(3, 1, 2)
plt.plot(frequencia, bateria, color='grey')
plt.title('Espectrografia da Bateria')
plt.ylabel('Amplitude')

# Contrabaixo
plt.subplot(3, 1, 3)
plt.plot(frequencia, contrabaixo, color='brown')
plt.title('Espectrografia do Contrabaixo')
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Amplitude')

# Mostrar o grfico
plt.tight_layout()
plt.show()
```
